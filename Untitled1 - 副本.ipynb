```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.preprocessing import StandardScaler

# ========== 1  ==========
df = pd.read_excel('BP-qiti.xlsx')
features = ['Eads', 'Q', 'd', 'p', 'VBM', 'CBM', 'Ef', 'Eg', 'WF']
X = df[features]
y = df['Response']

# 
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# ========== 2  ==========
model = ExtraTreesClassifier(n_estimators=200, random_state=42)
model.fit(X_scaled, y)

# ========== 3  ==========
feat1, feat2 = 'Eads', 'p'
i1, i2 = features.index(feat1), features.index(feat2)

# 
mean_values = X_scaled.mean(axis=0)

# ========== 4  ==========
x_min, x_max = X_scaled[:, i1].min() - 0.5, X_scaled[:, i1].max() + 0.5
y_min, y_max = X_scaled[:, i2].min() - 0.5, X_scaled[:, i2].max() + 0.5

xx, yy = np.meshgrid(np.linspace(x_min, x_max, 300),
                     np.linspace(y_min, y_max, 300))

# 
grid_points = np.tile(mean_values, (xx.size, 1))
grid_points[:, i1] = xx.ravel()
grid_points[:, i2] = yy.ravel()

# ========== 5  ==========
Z = model.predict(grid_points)
Z = Z.reshape(xx.shape)

# ========== 6  ==========
plt.figure(figsize=(8, 6))
plt.contourf(xx, yy, Z, cmap='coolwarm', alpha=0.4)
plt.scatter(X_scaled[(y.values==1), i1], X_scaled[(y.values==1), i2], label='Response', color='red', edgecolor='k')
plt.scatter(X_scaled[(y.values==0), i1], X_scaled[(y.values==0), i2], label='Non-Response', color='blue', edgecolor='k')

plt.xlabel(f'{feat1} (scaled)', fontsize=14)
plt.ylabel(f'{feat2} (scaled)', fontsize=14)
plt.legend(fontsize=12)
plt.title('Decision Boundary of Extra Trees', fontsize=15)
plt.tight_layout()
plt.show()
```

```python
# PDP  ICE
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import NearestNeighbors

# 
GRID = 120        #  220 -> 120 3.3x
K_NEIGHBORS = 10  #  20 -> 10
DO_ICE = False    #  ICE PDP True

# X, X_scaled, y, model, scaler, features, feat1, feat2, i1, i2
x_min, x_max = X_scaled[:, i1].min() - 0.5, X_scaled[:, i1].max() + 0.5
y_min, y_max = X_scaled[:, i2].min() - 0.5, X_scaled[:, i2].max() + 0.5
xx, yy = np.meshgrid(np.linspace(x_min, x_max, GRID), np.linspace(y_min, y_max, GRID))

# 1) PDP/
base = X_scaled  # (n, p)
Zp_pdp = np.zeros_like(xx, dtype=float)
for r in range(xx.shape[0]):
    grid_points = base.copy()
    #  Eads, Ep -> 
    grid_points[:, i1] = xx[r, 0]
    grid_points[:, i2] = yy[r, 0]
    Zp_pdp[r, :] = model.predict_proba(grid_points)[:, 1].mean()

# 2)  ICE
if DO_ICE:
    k = min(K_NEIGHBORS, len(X_scaled))
    nbrs = NearestNeighbors(n_neighbors=k).fit(X_scaled[:, [i1, i2]])
    Zp_ice = np.zeros_like(xx, dtype=float)
    # 
    grid_2d = np.c_[xx.ravel(), yy.ravel()]
    dists, idxs = nbrs.kneighbors(grid_2d, return_distance=True)
    for j, (xc, yc) in enumerate(grid_2d):
        local = X_scaled[idxs[j]].copy()
        local[:, i1] = xc
        local[:, i2] = yc
        Zp_ice.ravel()[j] = model.predict_proba(local)[:, 1].mean()

# 
if DO_ICE:
    fig, axes = plt.subplots(1, 2, figsize=(13, 5), sharex=True, sharey=True)
    panels = [(axes[0], Zp_pdp, 'PDP decision surface (avg over others)'),
              (axes[1], Zp_ice, f'Local ICE surface (k={K_NEIGHBORS})')]
else:
    fig, ax = plt.subplots(1, 1, figsize=(8, 6))
    panels = [(ax, Zp_pdp, 'PDP decision surface (avg over others)')]

for ax, Zp, title in panels:
    cs = ax.contourf(xx, yy, Zp, levels=20, cmap='coolwarm', alpha=0.6)
    ax.contour(xx, yy, Zp, levels=[0.5], colors='k', linewidths=1.2)
    ax.scatter(X_scaled[y.values==1, i1], X_scaled[y.values==1, i2], c='red', edgecolor='k', s=35, label='Response')
    ax.scatter(X_scaled[y.values==0, i1], X_scaled[y.values==0, i2], c='blue', edgecolor='k', s=35, label='Non-Response')
    ax.set_xlabel(f'{feat1} (scaled)')
    ax.set_ylabel(f'{feat2} (scaled)')
    ax.set_title(title)

handles, labels = panels[0][0].get_legend_handles_labels()
fig.legend(handles, labels, loc='upper right')
fig.colorbar(cs, ax=[p[0] for p in panels], label='P(class=1)')
plt.tight_layout()
plt.show()

```

```python
#  SHAP 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
from itertools import combinations
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.pipeline import make_pipeline
import shap

# ==========  ==========
#  X, y, features 
try:
    _ = X.shape
    _ = y.shape
    _ = len(features)
except NameError:
    # 
    df = pd.read_excel('BP-qiti.xlsx')
    features = ['Eads', 'Q', 'd', 'p', 'VBM', 'CBM', 'Ef', 'Eg', 'WF']
    X = df[features]
    y = df['Response']

#  X_scaled, model, mean_values 
try:
    _ = X_scaled.shape
    _ = model
    _ = mean_values.shape
except NameError:
    # 
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)
    # 
    model = ExtraTreesClassifier(n_estimators=200, random_state=42)
    model.fit(X_scaled, y)
    # 
    mean_values = X_scaled.mean(axis=0)

#  SHAP  order
try:
    top_indices = order[:5]
    mean_abs = np.abs(vals).mean(axis=0) if 'vals' in globals() else None
except NameError:
    # SHAP
    explainer = shap.Explainer(model)
    shap_values = explainer(X_scaled)
    vals = shap_values.values
    if vals.ndim == 3:
        vals = vals[..., 1]
    mean_abs = np.abs(vals).mean(axis=0)
    order = np.argsort(-mean_abs)
    top_indices = order[:5]

TOP5 = [features[i] for i in top_indices]
print("TOP5 features:", TOP5)

#  Unicode  LaTeX 
display_map = {
    'Eads': r'$E_{ads}$',
    'Q': r'$\Delta Q$',
    'd': 'd',
    'p': r'$\epsilon_p$',
    'VBM': 'VBM',
    'CBM': 'CBM',
    'Ef': r'$E_f$',
    'Eg': r'$E_g$',
    'WF': r'$W_F$'
}

# ==========  ==========
# SHAPSHAP
if mean_abs is not None:
    # SHAP
    all_shap_values = mean_abs
else:
    # SHAP
    explainer_tmp = shap.Explainer(model)
    shap_values_tmp = explainer_tmp(X_scaled)
    vals_tmp = shap_values_tmp.values
    if vals_tmp.ndim == 3:
        vals_tmp = vals_tmp[..., 1]
    all_shap_values = np.abs(vals_tmp).mean(axis=0)

# 91
total_importance = all_shap_values.sum()
normalized_importance = all_shap_values / total_importance

# TOP5
top5_normalized = normalized_importance[top_indices]
top5_labels = [display_map.get(f, f) for f in TOP5]

# 
plt.figure(figsize=(10, 6))
# Arial
plt.rcParams['font.family'] = 'Arial'
plt.rcParams['font.sans-serif'] = ['Arial']

# 
import matplotlib as mpl
cmap = mpl.colormaps.get_cmap('Blues') if hasattr(mpl, 'colormaps') else mpl.cm.get_cmap('Blues')
colors = [cmap(0.4 + 0.5*i/max(1, len(top5_normalized)-1)) for i in range(len(top5_normalized))]
bars = plt.bar(range(len(top5_normalized)), top5_normalized, color=colors, edgecolor='#1f4e79', linewidth=1.5)

# 24
plt.xticks(range(len(top5_normalized)), top5_labels, fontsize=24, fontfamily='Arial')
plt.ylabel('Feature Importance (Normalized)', fontsize=24, fontfamily='Arial')
plt.title('Top-5 Features Importance (Sum of All 9 Features = 1.0)', fontsize=24, fontweight='bold', fontfamily='Arial', pad=20)

# 
plt.grid(False)

# 1.5
ax = plt.gca()
for spine in ax.spines.values():
    spine.set_linewidth(1.5)

# y
max_val = max(top5_normalized)
plt.ylim(0, max_val * 1.15)  # 15%

# 
for i, (bar, val) in enumerate(zip(bars, top5_normalized)):
    # 
    label_y = val + max_val * 0.02  # 
    plt.text(bar.get_x() + bar.get_width()/2, label_y, 
             f'{val*100:.2f}%', ha='center', va='bottom', fontsize=24, fontweight='bold', fontfamily='Arial')

# 
ax.tick_params(axis='both', which='major', length=6, width=1.5, labelsize=24)
ax.tick_params(axis='both', which='minor', length=4, width=1.5)
plt.tight_layout()
# 
shap_bar_path = os.path.join('figures-1103', 'top5_shap_barplot.png')
plt.savefig(shap_bar_path, dpi=220, bbox_inches='tight')
plt.show()

# 
print(f"\n9 = {normalized_importance.sum():.6f}")
print(f"TOP5 = {top5_normalized.sum():.6f}")
print(f"TOP5 = {top5_normalized.sum()*100:.2f}%")

# 
pairs = list(combinations(TOP5, 2))
print("Total pairs:", len(pairs), pairs)

# 
import os
os.makedirs('figures-1103', exist_ok=True)

# 
GRID = 200

# 
pair_acc = []

for (f1, f2) in pairs:
    i1, i2 = features.index(f1), features.index(f2)

    # 1)  model
    x_min, x_max = X_scaled[:, i1].min() - 0.5, X_scaled[:, i1].max() + 0.5
    y_min, y_max = X_scaled[:, i2].min() - 0.5, X_scaled[:, i2].max() + 0.5
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, GRID),
                         np.linspace(y_min, y_max, GRID))
    grid_points = np.tile(mean_values, (xx.size, 1))
    grid_points[:, i1] = xx.ravel()
    grid_points[:, i2] = yy.ravel()
    Z = model.predict(grid_points).reshape(xx.shape)

    # 2) 5-fold CV
    Xi = X[[f1, f2]].values
    cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
    clf2 = make_pipeline(StandardScaler(with_mean=True, with_std=True),
                         ExtraTreesClassifier(n_estimators=300, random_state=42))
    acc = cross_val_score(clf2, Xi, y.values, cv=cv, scoring='accuracy').mean()
    pair_acc.append((f1, f2, float(acc)))

    # 3) 
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))
    
    # Arial
    plt.rcParams['font.family'] = 'Arial'
    plt.rcParams['font.sans-serif'] = ['Arial']

    # 
    ax = axes[0]
    ax.contourf(xx, yy, Z, cmap='coolwarm', alpha=0.4)
    ax.scatter(X_scaled[(y.values==1), i1], X_scaled[(y.values==1), i2], label='Response', color='red', edgecolor='k', s=25)
    ax.scatter(X_scaled[(y.values==0), i1], X_scaled[(y.values==0), i2], label='Non-Response', color='#2E86AB', edgecolor='k', s=25)
    ax.set_xlabel(f'{display_map.get(f1, f1)} (scaled)', fontsize=24, fontfamily='Arial')
    ax.set_ylabel(f'{display_map.get(f2, f2)} (scaled)', fontsize=24, fontfamily='Arial')
    ax.set_title('Decision Boundary', fontsize=24, fontweight='bold', fontfamily='Arial')
    ax.legend(loc='upper right', fontsize=24, prop={'family': 'Arial'})
    # 1.5
    for spine in ax.spines.values():
        spine.set_linewidth(1.5)
    # 
    ax.tick_params(axis='both', which='major', length=6, width=1.5, labelsize=24)
    ax.tick_params(axis='both', which='minor', length=4, width=1.5)

    # 
    ax2 = axes[1]
    bar_color = '#4C78A8'
    ax2.bar([f'{display_map.get(f1, f1)} + {display_map.get(f2, f2)}'], [acc], color=bar_color, edgecolor='#1f4e79', linewidth=1.5)
    ax2.set_ylim(0.0, 1.05)
    ax2.set_ylabel('CV Accuracy (5-fold)', fontsize=24, fontfamily='Arial')
    ax2.set_title('Accuracy', fontsize=24, fontweight='bold', fontfamily='Arial')
    for p, v in zip(ax2.patches, [acc]):
        ax2.text(p.get_x()+p.get_width()/2, v-0.05, f'{v:.3f}', ha='center', va='top', fontsize=24, color='white', fontweight='bold', fontfamily='Arial')
    # 1.5
    for spine in ax2.spines.values():
        spine.set_linewidth(1.5)
    # 
    ax2.tick_params(axis='both', which='major', length=6, width=1.5, labelsize=24)
    ax2.tick_params(axis='both', which='minor', length=4, width=1.5)

    fig.suptitle(f'Pair: {display_map.get(f1, f1)} + {display_map.get(f2, f2)}', fontsize=24, fontweight='bold', fontfamily='Arial')
    fig.tight_layout()
    out_path = os.path.join('figures-1103', f'decision_boundary_accuracy_{f1}_{f2}.png')
    fig.savefig(out_path, dpi=200, bbox_inches='tight')
    plt.show()

# 4) 10 
pair_acc_sorted = sorted(pair_acc, key=lambda x: x[2], reverse=True)
labels = [f'{display_map.get(a, a)} + {display_map.get(b, b)}' for a, b, _ in pair_acc_sorted]
values = [v for _, _, v in pair_acc_sorted]

plt.figure(figsize=(20, 10))  # 
# Arial
plt.rcParams['font.family'] = 'Arial'
plt.rcParams['font.sans-serif'] = ['Arial']

# 
import matplotlib as mpl
cmap = mpl.colormaps.get_cmap('Blues') if hasattr(mpl, 'colormaps') else mpl.cm.get_cmap('Blues')
colors = [cmap(0.45 + 0.5*i/max(1, len(values)-1)) for i in range(len(values))]

# 
bar_width = 0.6  # 
x_positions = range(len(values))
bars = plt.bar(x_positions, values, width=bar_width, color=colors, edgecolor='#1f4e79', linewidth=1.5)

plt.xticks(x_positions, labels, rotation=35, ha='right', fontsize=24, fontfamily='Arial')
# y
max_val = max(values)
plt.ylim(0.0, max_val * 1.20)  # 20%
plt.ylabel('CV Accuracy (5-fold)', fontsize=24, fontfamily='Arial')
plt.title('Top-5 pair combinations accuracy ranking', fontsize=24, fontweight='bold', fontfamily='Arial', pad=20)

# 
for i, (bar, v) in enumerate(zip(bars, values)):
    # 
    label_y = v + max_val * 0.03  # 3%
    plt.text(bar.get_x() + bar.get_width()/2, label_y, 
             f'{v:.3f}', ha='center', va='bottom', fontsize=24, color='black', 
             fontweight='bold', fontfamily='Arial')

# 1.5
ax = plt.gca()
for spine in ax.spines.values():
    spine.set_linewidth(1.5)
# 
ax.tick_params(axis='both', which='major', length=6, width=1.5, labelsize=24)
ax.tick_params(axis='both', which='minor', length=4, width=1.5)

plt.tight_layout()
rank_path = os.path.join('figures-1103', 'pair_accuracy_ranking.png')
plt.savefig(rank_path, dpi=220, bbox_inches='tight')
plt.show()

```
